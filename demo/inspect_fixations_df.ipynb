```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from eyelinkio import read_edf
import numpy as np
```

```python
# enter file path
filepath = '/Users/lydia/visualEyes/data/TG_2015.EDF'
```

```python
# read the cdf
edf_file = read_edf(filepath)
```

```python
df = edf_file.to_pandas()
```

```python
fixation_duration = []

for index, series in df['discrete']['fixations'].iterrows():
    single_duration = series['etime'] - series['stime']
    fixation_duration.append(single_duration)

print(round(max(fixation_duration), 2))
```
